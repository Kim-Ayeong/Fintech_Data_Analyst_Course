### 데이터를 API로 가져와 본다 - 공공 데이터

REST API 방식으로 데이터를 불러 오도록 한다. 먼저 다음과 같은 절차로 신청한다.<br>

1. 먼저 다음 사이트에 접속한다: www.data.go.kr 
2. 사이트에 가입하고 로그인 한다.
3. 상단 탭에서: 데이터 셋 => 오픈 API로 들어간다
4. "상가(상권) 정보"를 클릭하여 들아간다.
5. "활용신청" 버튼을 클릭하고 신청 양식을 작성한다.
   - "활용목적"은 "참고자료"로 선택해 둔다.
   - "상세기능정보"는 모두 체크 선택해 둔다.
   - "동의합니다"에 체크를 해 둔다.
6. "신청" 버튼을 클릭해서 신청 완료. 

신청이 완료된 후 다음 순서로 인증키를 받아 두어야만 한다. <br>
 
1. 마이페이지 => 오픈 API => 개발계정 => 방금 승인된 아이템을 통해서 "개발계정 상세보기"로 들어간다.
2. "일반 인증키" 발급을 클릭해서 인증키를 발급 받는다.
3. 인증키를 복사해 둔다.

#### (추가) 지번단위 상가업소 조회 및 파일 저장

```python
import pandas as pd
```

```python
parameterset = "/baroApi?resId=store&catId=pnu&key=1123010300207370002&ServiceKey=" + serviceKey
url = endpoint + parameterset
res = rq.get(url)
print("Status = " + str(res.status_code))
```

```python
res_text = res.content # = res.text
soup = bs4.BeautifulSoup(res_text, 'lxml') # = html.parser
print(soup.prettify())
```

```python
col_kr = soup.find('columns').text.split(',')
col_kr
```

```python
col_en = [c.name for c in soup.find('item').children if c != '\n']
col_en
```

```python
startnumber=1
endnumber=1000

data = {}
```

```python
for c_kr in col_kr:
    data[c_kr] = []
data
```

```python
for c_kr, c_en in zip(col_kr, col_en):
    item_list = []
    
    for x in soup.find_all(c_en):
        item_list.append(x.text)
        
    data[c_kr] = item_list
```

```python
dataframe = pd.DataFrame(data)
dataframe
```

```python
dataframe.to_csv('ex.csv')
```
