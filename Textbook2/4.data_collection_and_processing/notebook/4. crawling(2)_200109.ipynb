### 실제 사이트 크롤링 - 네이버 뉴스

```python
import bs4
import re
import requests as rq
```

#### 사이트에 접속하고 탐색해 본다:

1. 먼저 다음 사이트에 들어가서 살펴본다:  [네이버 뉴스](https://news.naver.com)
2. 크롬에서 3점 메뉴 버튼을 누르고 **도구 더보기** ==> **개발자 도구**를 선택해서 개발자 도구를 연다.
3. 개발자 도구 왼쪽 상단의 화살표 버튼을 눌러서 사이트 일부를 서택하고 해당 HTML코드를 볼 수 있다. "[CTRL] + [SHIFT] + C" 조합으로 대체할 수 있다.

```python
# 웹사이트의 URL에 접속.
res = rq.get("https://news.naver.com")
```

```python
# status_code가 200이면 OK.
# status_code가 4xx이면 접속 오류.
res.status_code
```

```python
# 데이터 전체를 출력해 본다.
# 아직은 이해하기 어렵다.
print(res.text)              # print(res.content)와 비교.
```

```python
my_html = res.text
soup = bs4.BeautifulSoup(my_html, 'html.parser')     # BeautifulSoup 객체 반환. 'html.parser' 사용.
print(soup.prettify())  
```

```python
# 모든 div 태그의 attr을 출력해 본다. 
x=soup.find_all('div')
n = len(x)
for i in range(n):
    print(x[i].get('class'))
```

#### 헤드라인 가져오기:

먼저 ul 태그로 정리된 헤드라인 리스트 클래스를 가져와 본다.

```python
res = soup.find_all('ul',class_='hdline_article_list')
for x in res:
    print(x.contents)
```

위에서 보았을 때 div 태그안에 있는 헤드라인 기사 타이틀을 가져와 본다.

```python
res = soup.find_all('div',class_='hdline_article_tit')
for x in res:
    print(x.text.strip())
```

#### 링크 타이틀 중에서 볼드 처리 뉴스 타이틀과 하이퍼링크 가져오기:

```python
res_a = soup.find_all('a', class_=re.compile('clicks'))      # "clicks" 패턴 있는 클래스.
n_count = 0
news = []
for x in res_a:
    res = x.find('strong')                                   # 다시 검색.
    if res:                                                  # None이 아닌 경우만.
        an_article = res.text
        if (len(an_article) > 10):                           # 길이가 10 이상인 경우만.
            n_count += 1
            print(n_count, end = " : ")
            print(an_article, end = " => ")
            print(x.get('href'))
            news += [(an_article, x.get("href"))]
```

#### 하이퍼링크 따라서 뉴스 기사 가져오기:

```python
for news_title, hlink in news:
        res = rq.get(hlink)
        my_html = res.text
        soup = bs4.BeautifulSoup(my_html, 'html.parser')  
        body = soup.find(id='articleBodyContents')
        if (body):                                           # None이 아닌 경우만 출력.
            print(news_title, end=" : ")
            print(body.text.strip()[63:])                    # 서두의 불필요 부분 제거.      
            print("-"*100+"\n")
```
